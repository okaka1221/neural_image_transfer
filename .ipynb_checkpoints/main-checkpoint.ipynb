```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.figsize'] = (12,12)
mpl.rcParams['axes.grid'] = False
import time
import IPython.display as display
import PIL.Image

import tensorflow as tf
from keras.preprocessing import image
```

```python
content_path = './FlindersStStation.jpg'
style_path = './StarryNight.jpg'

width, height = image.load_img(content_path).size
size = [height, width]

content_layers = ['block4_conv2']
style_layers = ['block1_conv1', 'block2_conv1', 'block3_conv1', 'block4_conv1', 'block5_conv1']
num_content_layers = len(content_layers)
num_style_layers = len(style_layers)

content_weight = 1e3
style_weight = 1
```

```python
def load_img(input_path, size):
    img = image.load_img(input_path, target_size=size)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img /= 255
    return img
```

```python
def imshow(img, tensor=False):
    if tensor:
        img = img.numpy()
    img = np.squeeze(img, axis=0)
    img *= 255
    img = img.astype(np.uint8)
    plt.imshow(img)
```

```python
def tensor_to_img(tensor):
    tensor = tensor*255
    tensor = np.array(tensor, dtype=np.uint8)
    if np.ndim(tensor)>3:
        assert tensor.shape[0] == 1
        tensor = tensor[0]
    return PIL.Image.fromarray(tensor)
```

```python
noise_img = np.random.uniform(-20, 20, (1, size[0], size[1], 3)).astype(np.float32) / 255
content_img = load_img(content_path, size)
style_img = load_img(style_path, size)
```

```python
imshow(content_img)
```

```python
imshow(style_img)
```

```python
def get_vgg_model(layer_names):
    vgg = tf.keras.applications.vgg19.VGG19(include_top=False, weights='imagenet')
    vgg.trainable = False

    outputs = [vgg.get_layer(name).output for name in layer_names]
    model = tf.keras.Model([vgg.input], outputs)
    return model
```

```python
def gram_matrix(input_tensor):
    result = tf.linalg.einsum('bijc,bijd->bcd', input_tensor, input_tensor)
    input_shape = tf.shape(input_tensor)
    num_locations = tf.cast(input_shape[1] * input_shape[2], tf.float32)
    return result / num_locations
```

```python
def get_representations(img):
    vgg = get_vgg_model(content_layers + style_layers)
    preprocessed_input = tf.keras.applications.vgg19.preprocess_input(img * 255)
    outputs = vgg(preprocessed_input)
    
    content_outputs = outputs[:num_content_layers]
    style_outputs = outputs[num_content_layers:]
    style_outputs = [gram_matrix(style_output) for style_output in style_outputs]

    content_features = {content_name: value for content_name, value in zip(content_layers, content_outputs)}
    style_features = {style_name: value for style_name, value in zip(style_layers, style_outputs)}
                      
    return (content_features, style_features)
```

```python
def clip_0_1(img):
    return tf.clip_by_value(img, clip_value_min=0., clip_value_max=1.0)
```

```python
def get_loss(outputs, content_targets, style_targets):
    content_outputs = outputs[0]
    style_outputs = outputs[1]
    
    content_loss = tf.add_n([tf.reduce_mean((content_outputs[name]-content_targets[name])**2) for name in content_outputs.keys()])
    content_loss *= content_weight / num_content_layers
    
    style_loss = tf.add_n([tf.reduce_mean((style_outputs[name]-style_targets[name])**2) for name in style_outputs.keys()])
    style_loss *= style_weight / num_style_layers
    
    loss = content_loss + style_loss
    return loss
```

```python
content_targets = get_representations(content_img)[0]
style_targets = get_representations(style_img)[1]

img = tf.Variable(content_img)
opt = tf.optimizers.Adam(learning_rate=0.02, beta_1=0.99, epsilon=1e-1)
```

```python
def train_step(img):
    with tf.GradientTape() as tape:
        outputs = get_representations(img)
        loss = get_loss(outputs, content_targets, style_targets)
    
    grad = tape.gradient(loss, img)
    opt.apply_gradients([(grad, img)])
    img.assign(clip_0_1(img))
```

```python
start = time.time()

n_iteration = 3000

step = 0
for n in range(n_iteration):
    step += 1
    train_step(img)
    if step % 50 ==0:
        display.clear_output(wait=True)
        display.display(tensor_to_img(img))
        tensor_to_img(img).save('output_' + str(step) + '.jpg')
        print("Train step: {}".format(step))
    
end = time.time()
print("Total time: {:.1f}".format(end-start))
```
